```python
import numpy as np
import matplotlib.pyplot as plt
import math 

def f(x):
    return np.sin(x) + 0.5 * x
x = np.linspace(-2* np.pi, 2* np.pi, 50)

plt.plot(x, f(x), 'b')
plt.grid(True)
```

```python
reg = np.polyfit(x, f(x), deg=1)
ry = np.polyval(reg, x)
```

```python
plt.plot(x,f(x),'b', label='f(x)')
plt.plot(x,ry, 'r', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```
